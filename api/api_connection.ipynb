```python
API_URL = 'https://api.egytech.fyi/participants'
```

```python
import requests
import pandas as pd

try:
    # Make a GET request to the API endpoint
    response = requests.get(API_URL)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        # Extract the population data from the response
        participants_data = data['results']
        
        # Create a DataFrame from the population data
        df = pd.DataFrame(participants_data)
        #print(df)
        # make the DataFrame pretty 
        
        
        # this lines to make the dataframe more readable and styling it  to working with it in jupyter notebook for some researching points 
        # df = df.style.set_properties(**{'text-align': 'left'})
        # df = df.set_table_styles([dict(selector='th', props=[('text-align', 'left')])])
        display(df)
        # Save the DataFrame to a CSV file
        #df.to_csv('../data/raw/v0/EgyTech_participants_2024_v0.csv', index=False)
    else:
        # Handle the case where the request was not successful
        print("Error:", response.status_code, response.text)
except Exception as e:
    # Handle any exceptions that occur during the request
    print("Error:", str(e))
```
